# Energy demand EDA

Seasonality, weather response and lag behavior for the generated hourly demand series.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(str(Path.cwd().parent))
from train import add_lag_features, make_energy_series

sns.set_theme(style="whitegrid")
df = make_energy_series(days=730, random_state=31)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
df.set_index("timestamp")["demand_mw"].iloc[:24 * 28].plot(ax=ax)
ax.set_title("First four weeks of hourly demand")
ax.set_ylabel("MW")
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.lineplot(data=df, x="hour", y="demand_mw", errorbar=None, ax=axes[0])
sns.lineplot(data=df, x="month", y="demand_mw", errorbar=None, ax=axes[1])
axes[0].set_title("Average demand by hour")
axes[1].set_title("Average demand by month")
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.scatterplot(data=df.sample(2000, random_state=31), x="temperature_c", y="demand_mw", hue="is_weekend", alpha=0.5, ax=axes[0])
sns.boxplot(data=df, x="is_weekend", y="demand_mw", ax=axes[1])
axes[0].set_title("Demand vs temperature")
axes[1].set_title("Weekend effect")
plt.tight_layout()
```

```python
features = add_lag_features(df)
cols = ["demand_mw", "lag_1", "lag_24", "lag_168", "rolling_mean_24", "temperature_c"]
sns.heatmap(features[cols].corr(), annot=True, cmap="vlag", center=0)
plt.title("Lag and weather correlation")
plt.tight_layout()
```
